```python
# ===============================
# 1. Import libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set(style="whitegrid", font_scale=1.2)
```

```python
# ===============================
# 2. Load dataset
# ===============================
df = pd.read_csv("student_habits_performance.csv")

# Show first rows
df.head()
```

```python
# ===============================
# 3. Basic info & summary statistics
# ===============================
print("===== Dataset Info =====")
df.info()

print("\n===== Summary Statistics =====")
df.describe()
```

```python
# ===============================
# 4. Distribution plots (Histogram)
# ===============================
numeric_cols = [
    "study_hours_per_day",
    "social_media_hours",
    "netflix_hours",
    "sleep_hours",
    "exercise_frequency",
    "attendance_percentage",
    "mental_health_rating",
    "exam_score"
]

df[numeric_cols].hist(figsize=(14, 10), bins=20, color='skyblue')
plt.suptitle("Distribution of Numeric Features", fontsize=18)
plt.show()
```

```python
# ===============================
# 5. Boxplots (optional, helpful for outliers)
# ===============================
plt.figure(figsize=(14, 8))
df[numeric_cols].boxplot()
plt.title("Boxplot of Numeric Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# ===============================
# 6. Correlation Matrix
# ===============================
corr = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
```
